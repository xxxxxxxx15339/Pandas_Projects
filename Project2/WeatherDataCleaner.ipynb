```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Loading the CSV file into the DataFrame :


```python
df = pd.read_csv('weather_data.csv')
```

# Clean missing temperature values by replacing them with the mean temperature :


```python
# Calculer le Mean and the number of rows :
Temp_Mean = df['Temperature'].mean(axis=0)
num_rows = df.shape[0]

print(df)

# remplacer NaN avec le Mean
for i in range(0, num_rows):
    if pd.isnull(df.at[i, 'Temperature']):
        df.at[i, 'Temperature'] = Temp_Mean

df = df.round(1)
```

# Filter data for July only :

```python
# Make sure your date column is datetime :
df['Date'] = pd.to_datetime(df['Date'])

# Remove all the other months and keep only July :
df_july = df[df['Date'].dt.month == 7]
print(df_july)
```

# Calculate the average temperature in July.

```python
# Calculate the average temperature in July :
Avg_July = df_july['Temperature'].mean()
print(Avg_July)
```

# Plot daily temperatures for July :

```python
plt.figure(figsize=(10, 5))
plt.plot(df_july['Date'], df_july['Temperature'], marker='o', linestyle='-', color='orange')

plt.title("ðŸ“ˆ Daily Temperatures in July")
plt.xlabel("Date")
plt.ylabel("Temperature (ÂºC)")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
